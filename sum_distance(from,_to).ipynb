```python
def sum_distance(start, end):
    if start > end:
        start, end = end, start
    total = 0
    for number in range(start, end + 1):
        total += number
    return total

# Тест
print(f"sum_distance(1, 5) = {sum_distance(1, 5)}")      # Выводится: 15
print(f"sum_distance(5, 1) = {sum_distance(5, 1)}")      # Выводится: 15 (меняет местами)
print(f"sum_distance(-3, 3) = {sum_distance(-3, 3)}")    # Выводится: 0
print(f"sum_distance(10, 10) = {sum_distance(10, 10)}")  # Выводится: 10
```
